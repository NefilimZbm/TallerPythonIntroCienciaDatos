## Preprocesamiento

En este ejemplo, lo datos no requieren de un proceso de limpieza salvo los que hemos realizado (eliminar las ultimas columnas) asi que dejaremos continuaremos con la transformacion de los datos.

### Datos de entrenamiento

A los datos de entrenamiento se les aplica el siguiente proceso:

1. extraccion de sensores con valores efectivos
2. etiquetado de la condicion en 4 categorias con base en los ciclos restantes hasta que ocurre la falla

```
# 1. seleccion de los sensores con lecturas efectivas
dfTest_orig.describe()
```

```
# selecciona los valores de la desviacion estandar de las columnas
dfTest_orig.describe().loc["std",:]
```

Podemos inferir que si la desviacion estandar de una columna es igual a 0. entonces los valores de esta columna no estan cambiando, por lo tanto no ofrecen ningun tipo de informacion del sistema:

```
dfTest_orig.loc[:,"s1"].plot()
```

Vamos a elegir solo aquellas columnas que tienen una desviacion estandard distinta de cero.

```
# concatenate columns with index in [0:2, 6:9, 10:14, 15:20, 21, 24:26]
# dfTrain ya tienen solo los sensores con valores efectivos
dfTrain = pd.concat([
    dfTrain_orig.iloc[:,0:2], dfTrain_orig.iloc[:,6:9],
    dfTrain_orig.iloc[:,10:14], dfTrain_orig.iloc[:,15:20],
    dfTrain_orig.iloc[:,21], dfTrain_orig.iloc[:,24:26]], axis=1)
```

```
# podemos revisar que ninguna columna tienen desv. stand. nula
dfTrain.describe().loc["std",:]
```

Ahora vamos a etiquetar las columnas de acuerdo al ultimo ciclo que se tiene registrado para cada turbina. El siguiente codigo agrupa los datos por `id`, selecciona la columna `cycle` de cada grupo y determina el valor maximo:

```
df_maxCycle = dfTrain.groupby('id')['cycle'].max()
```

```
df_maxCycle.columns = ["id", "last_cycle"]
```

```
type(df_maxCycle)
```

```
df_maxCycle
```

La serie anterior significa que la turbina con `id` 1 opera durante 192 ciclos antes de fallar y la turbina con `id` 100 opera durante 200 ciclos antes de fallar. Esta columna corresponde al nombre `ttf` que usamos para los datos de verificacion: "ttf" = _time to failure_. Para agregarla a nuestros datos como una columna que indique el tiempo restante antes de falla necesitamos convertirla en un `DataFrame`. Lo rehacemos de la siguiente forma

```
df_maxCycle = pd.DataFrame(dfTrain.groupby('id')['cycle'].max())
df_maxCycle.reset_index(level=0,inplace=True)
df_maxCycle.columns = ['id', 'last_cycle']
```

```
# agregamos la columna usando el `id` para llenarla con el ciclo maximo  
dfTrain = pd.merge(dfTrain, df_maxCycle, on='id')

# construimos una nueva columna que es la diferencia y la llamamos ttf
dfTrain['ttf'] = dfTrain['last_cycle'] - dfTrain['cycle']

# eliminamos la columna del ultimo cicle de cada turbina pues ya no se requiere
dfTrain.drop(['last_cycle'], axis=1, inplace=True)
```

A continuacion vamos a crear las etiquetas de clasificacion para cada renglon del `dataframe` con informacion categorica determinada a partir del numero de ciclos restantes antes que ocurra la falla.

Primero para clasificacion binaria dado una ventana de tiempo determinada por 30 ciclos



```
win = 30
dfTrain['ttf'].apply(lambda x: 1 if x<=win else 0)
```

Una forma elegante de hacelo es usando `qcut`

```
pd.qcut(dfTrain['ttf'], 2, labels=range(2))
```

```
# agregar columna con la etiqueta para clasificacion binaria bnc
dfTrain['label_bnc'] = dfTrain['ttf'].apply(lambda x: 1 if x<=win else 0)
```

Ahora para clasificacion multiclases de acuerdo a las siguientes categorias:

- de   0 - 50  ciclos restantes : urgente (1)
- de  51 - 125 ciclos restantes : corto   (2)
- de 126 - 200 ciclos restantes : medio   (3)
- de 201 -     ciclos restantes : largo   (4)


```
 # columna con etiquetas para clasificacion con multiples clases
  dfTrain['label_mcc'] = pd.qcut(dfTrain['ttf'], 4, labels=range(1, 5))
```

### Reduccion de dimensionalidad

Aplicamos PCA a los datos normalizados 

```
from sklearn.decomposition import PCA
```

```
dfTrainZ.columns[2:-3]
```

```
Xt = dfTrainZ.iloc[:,2:-3]
```

```
pca = PCA(n_components = 15)

pca.fit(Xt)

variance = pca.explained_variance_ratio_

expl = np.round(variance, decimals=3)*100

var = np.cumsum(expl)
```

```
plt.ylabel('% Variance Explained')
plt.xlabel('# of Features')
plt.title('PCA Analysis')
plt.ylim(0,100.5)
plt.style.context('seaborn-whitegrid')


plt.plot(expl)
plt.plot(var)
```

See how instead of using all the 14 attributes we can choose the first 2 principal components which explain almos 75% of the data

```
print(pca.explained_variance_ratio_)

print(pca.singular_values_)
```
